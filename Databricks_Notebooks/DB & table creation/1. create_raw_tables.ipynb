```sql
CREATE DATABASE if NOT EXISTS f1_raw
```

Create circuits table

```sql
DROP TABLE IF EXISTS f1_raw.circuits;
CREATE TABLE IF NOT EXISTS f1_raw.circuits (
  circuitId INT, 
  circuitRef STRING, 
  name STRING, 
  location STRING,
  country STRING,
  lat DOUBLE,
  lng DOUBLE,
  alt INT,
  url STRING
)
USING csv
OPTIONS (path "/mnt/formula1dlstudy/raw/circuits.csv", header "true")
```

```sql
SELECT * FROM f1_raw.circuits
```

Create races table

```sql
DROP TABLE IF EXISTS f1_raw.races;
CREATE TABLE IF NOT EXISTS f1_raw.races (
  raceId INT,
  year INT, 
  round INT, 
  circuitId INT,
  name STRING,
  date STRING,
  time STRING,
  url STRING
)
USING csv
OPTIONS (path "/mnt/formula1dlstudy/raw/races.csv", header "true")
```

```sql
SELECT * FROM f1_raw.races
```

Create constructors table

```sql
DROP TABLE IF EXISTS f1_raw.constructors;
CREATE TABLE IF NOT EXISTS f1_raw.constructors (
  constructorId INT,
  constructorRef STRING,
  name STRING,
  nationality STRING,
  url STRING
)
USING json
OPTIONS (path "/mnt/formula1dlstudy/raw/constructors.json")
```

```sql
SELECT * FROM f1_raw.constructors
```

Create drivers table

```sql
DROP TABLE IF EXISTS f1_raw.drivers;
CREATE TABLE IF NOT EXISTS f1_raw.drivers (
  driverId INT,
  driverRef STRING,
  number INT,
  code STRING,
  name STRUCT <forename: STRING, surname: STRING>,
  dob STRING,
  nationality STRING,
  url STRING
)
USING json
OPTIONS (path "/mnt/formula1dlstudy/raw/drivers.json")
```

```sql
SELECT * FROM f1_raw.drivers
```

Create drivers table

```sql
DROP TABLE IF EXISTS f1_raw.results;
CREATE TABLE IF NOT EXISTS f1_raw.results (
  resultId INT,
  raceId INT,
  driverId INT,
  constructorId INT,
  number INT,
  grid INT,
  position INT,
  positionText STRING,
  positionOrder INT,
  points INT,
  laps INT,
  time STRING,
  milliseconds INT,
  fastestLap INT,
  rank INT,
  fastestLapTime STRING,
  fastestLapSpeed FLOAT,
  statusId STRING
)
USING json
OPTIONS (path "/mnt/formula1dlstudy/raw/results.json")
```

```sql
SELECT * FROM f1_raw.results
```

Create pitstop table

```sql
DROP TABLE IF EXISTS f1_raw.pitstops;
CREATE TABLE IF NOT EXISTS f1_raw.pitstops (
raceId INT,
driverId INT,
duration STRING,
milliseconds INT,
lap INT,
stop INT,
time STRING
)
USING json
OPTIONS (path "/mnt/formula1dlstudy/raw/pit_stops.json",multiLine "true")
```

```sql
SELECT * FROM f1_raw.pitstops
```
